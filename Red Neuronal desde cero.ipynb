## Ahora con una red Neurona

```python
import numpy as np

# Clase de la red neuronal simple
class SimpleNeuralNetwork:
    def __init__(self):
        # Inicializar peso y sesgo aleatoriamente
        self.w = np.random.rand()
        self.b = np.random.rand()

    def forward(self, x):
        # Calcular la salida
        return self.w * x + self.b

    def train(self, X, y_true, learning_rate=0.01, epochs=1000):
        # Entrenamiento de la red neuronal
        for epoch in range(epochs):
            # PredicciÃ³n
            y_pred = self.forward(X)

            # CÃ¡lculo de la pÃ©rdida (error cuadrÃ¡tico medio)
            loss = np.mean((y_pred - y_true) ** 2)

            # CÃ¡lculo de los gradientes
            dw = (2 / len(X)) * np.dot(X, (y_pred - y_true))  # Derivada respecto a w
            db = (2 / len(X)) * np.sum(y_pred - y_true)      # Derivada respecto a b

            # ActualizaciÃ³n de los parÃ¡metros
            self.w -= learning_rate * dw
            self.b -= learning_rate * db

            # Opcional: imprimir la pÃ©rdida cada 100 Ã©pocas
            if epoch % 100 == 0:
                print(f'Ã‰poca {epoch}: PÃ©rdida = {loss}, w = {self.w}, b = {self.b}')

# Datos de ejemplo
X = dataset['Hours'].values
y_true = dataset['Scores'].values

# Inicializar la red neuronal
nn = SimpleNeuralNetwork()

# Entrenar la red neuronal
nn.train(X, y_true)

# Resultados finales
print(f"\nDespuÃ©s del entrenamiento: Peso (w) = {nn.w}, Sesgo (b) = {nn.b}")

```
