## Automatizacion de procesos


- Vamos a iniciar leyendo los archivos en una carpeta


```python
import os
import pandas as pd
```

```python
# Directorio donde se debe buscar los archivos
directorio_busqueda = r'E:\Wins\Arev\Pictures'

# Obtener la lista de archivos en el directorio
archivos = os.listdir(directorio_busqueda)

# Crear una lista para almacenar la información de cada archivo
data = []

for archivo in archivos:
    # Obtener la ruta completa del archivo
    ruta_archivo = os.path.join(directorio_busqueda, archivo)
    
    # Verificar si es un archivo (para excluir carpetas)
    if os.path.isfile(ruta_archivo):
        # Obtener el tamaño del archivo en bytes
        tamano = os.path.getsize(ruta_archivo)
        
        # Agregar la información del archivo a la lista
        data.append({'Nombre': archivo, 'Tamaño (bytes)': tamano})

# Crear un DataFrame con la información
df = pd.DataFrame(data)
```

```python
# Obtener la lista de archivos en el directorio
archivos = os.listdir(directorio_busqueda)

# Inicializar un contador para el consecutivo
contador = 1

for archivo in archivos:
    # Obtener la ruta completa del archivo
    ruta_actual = os.path.join(directorio_busqueda, archivo)
    
    # Verificar si es un archivo, comienza con "HPIM", y tiene la extensión .jpg
    if os.path.isfile(ruta_actual) and archivo.startswith('HPIM') and archivo.lower().endswith('.jpg'):
        # Crear el nuevo nombre con el formato deseado
        nuevo_nombre = f"Despedida Daniel-{contador}.jpg"
        
        # Obtener la ruta completa para el nuevo nombre
        ruta_nueva = os.path.join(directorio_busqueda, nuevo_nombre)
        
        # Renombrar el archivo
        os.rename(ruta_actual, ruta_nueva)
        
        # Incrementar el contador
        contador += 1

print(f"Se renombraron {contador - 1} archivos correctamente.")
```
